```python
import pandas as pd
import numpy as np
import os
```

# Create outcome table

## Load original csv

```python
outcome_file = "/home/tachennf/Documents/delta-rad/data/Madrid/clinical_data_clean.csv"
outcome_df = pd.read_csv(outcome_file, index_col=0, sep=";")
outcome_df.head()
```

## Transform to date format

```python
date_outcome_df = pd.DataFrame(index=outcome_df.index, columns=['date of last RT fraction', 'Date of local recurrence'])
date_outcome_df['date of last RT fraction'] = pd.to_datetime(outcome_df['date of last RT fraction'], format="%d/%m/%Y", errors='coerce')
date_outcome_df['Date of local recurrence'] = pd.to_datetime(outcome_df['Date of local recurrence'], format="%d/%m/%Y", errors='coerce')
date_outcome_df['Date last news'] = pd.to_datetime(outcome_df['Date last news'], format="%d/%m/%Y", errors='coerce')
date_outcome_df.head()
```

## Difference in date

```python
date_outcome_df['difference end RT and local recurrence'] = (date_outcome_df['Date of local recurrence'] - date_outcome_df['date of last RT fraction']).dt.days
nine_months_in_days = 365 
```

```python
date_outcome_df['one_year_local_control_status'] = date_outcome_df['difference end RT and local recurrence'] <= nine_months_in_days
date_outcome_df
```

```python
print(date_outcome_df['one_year_local_control_status'].value_counts())
```

```python
one_year_status_df = pd.DataFrame(index=date_outcome_df.index, columns=['one_year_local_control_status'])
one_year_status_df['one_year_local_control_status'] = date_outcome_df['one_year_local_control_status'].astype(int)
one_year_status_df
```

```python
one_year_status_df.to_csv('/home/tachennf/Documents/delta-rad/data/Madrid/combat_csv/one_year_local_control_status_df.csv', sep=';', index=True)
```
