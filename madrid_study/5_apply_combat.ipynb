```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from neuroCombat import neuroCombat
import os 
```

```python
madrid_path = '/home/tachennf/Documents/delta-rad/data/Madrid/filtered_csv/'
icm_path = '/home/tachennf/Documents/delta-rad/data/ICM_0.35T/extracted_radiomics/'
```

# Create joined tables
For all tables ...
Data must have participants as columns and features as rows
## Simu

```python
# columns = all participants
# rows = all features
madrid_simu = pd.read_csv(madrid_path+'simu_gtv.csv', index_col=0)
madrid_simu.head()
```

```python
icm_simu = pd.read_csv(icm_path+'simu_gtv.csv', index_col=0)
# icm_simu = icm_simu.transpose()
icm_simu.head()
```

### Join data

```python
# simu = pd.concat([madrid_simu, icm_simu], axis=1)
simu = pd.concat([madrid_simu, icm_simu], axis=0)
simu.head()
```

### Remove NaN before Combat

```python
simu = simu.dropna(axis=0, how='all')
simu.head()
```

```python
patients = simu.index.tolist()
batch_id = [0 if 'Mont' in patient else 1 for patient in patients] # 0: Madrid, 1: ICM
print(batch_id)
```

# Check distribution

### Plot

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.histplot(simu, x="original_glcm_Imc1", hue=batch_id, bins=30, kde=True)
plt.title('Histogram of original_glcm_Imc1')
plt.xlabel('original_glcm_Imc1')
plt.ylabel('Density')
plt.legend(title='Dataset', labels=['Madrid', 'ICM'])
```

### Kolgomorov-Smirnov test

```python
stats.kstest(icm_simu["original_glcm_Imc1"], madrid_simu["original_glcm_Imc1"], alternative='two-sided')
```

p-value is superior to 0.05 so distributions are different

# Apply Combat
## Reshape data

```python
madrid_simu = madrid_simu.transpose()
icm_simu = icm_simu.transpose()
simu = pd.concat([madrid_simu, icm_simu], axis=1)
simu = simu.dropna(axis=1, how='all')
simu.head()
```

```python
## ad batch_id row 
simu.loc['batch_id'] = batch_id
print(simu.shape)
```

```python
simu
```

## Harmonize data

```python
covars = {'batch_id': batch_id}
covars_df = pd.DataFrame(covars)
data_combat = neuroCombat(dat=simu, batch_col='batch_id', covars=covars_df)
```

```python
print(data_combat['data'].shape)
combat_df = pd.DataFrame(data_combat['data'], index=simu.index, columns=simu.columns)
```

# Check again distribution

### Plot

```python
combat_df = combat_df.transpose()
combat_df['batch_id'] = batch_id

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.histplot(combat_df, x="original_glcm_Imc1", hue=batch_id, bins=30, kde=True)
plt.title('Histogram of original_glcm_Imc1')
plt.xlabel('original_glcm_Imc1')
plt.ylabel('Density')
plt.legend(title='Dataset', labels=['Madrid', 'ICM'])
```

### Kolgomorov-Smirnov test

```python
# select patients for which batch_id is 0
combat_df_madrid = combat_df[combat_df['batch_id'] == 0]
combat_df_icm = combat_df[combat_df['batch_id'] == 1]

stats.kstest(combat_df_icm["original_glcm_Imc1"], combat_df_madrid["original_glcm_Imc1"], alternative='two-sided')
```

Now KS test says that statistics are the same (p = 0.02). 

# Apply Combat to all tables and features

### Madrid tables

```python
madrid_tables = os.listdir(madrid_path)
madrid_tables = sorted(madrid_tables)
print(madrid_tables)
```

### ICM tables

```python
icm_tables = os.listdir(icm_path)
icm_tables = sorted(icm_tables)
icm_tables.pop(-5)
```

```python
print(icm_tables)
```

## Application

```python
for table in icm_tables:

    icm_df = pd.read_csv(icm_path+table, index_col=0)
    icm_df = icm_df.transpose()
    icm_df = icm_df.dropna(axis=1, how='all')

    madrid_df = pd.read_csv(madrid_path+table, index_col=0)
    madrid_df = madrid_df.transpose()
    madrid_df = madrid_df.dropna(axis=1, how='all')

    data = pd.concat([madrid_df, icm_df], axis=1)
    patients = data.columns.tolist()

    batch_id = ["0" if 'Mont' in patient else "1" for patient in patients] # 0: Madrid, 1: ICM
    data.loc['batch_id'] = batch_id
        
    covars = {'batch_id': batch_id}
    covars_df = pd.DataFrame(covars)
    data_combat = neuroCombat(dat=data, batch_col='batch_id', covars=covars_df, ref_batch='1')
    
    combat_df = pd.DataFrame(data_combat['data'], index=data.index, columns=data.columns)
    patients = combat_df.columns.tolist()  
    batch_id = ['0' if 'Mont' in patient else '1' for patient in patients] # 0: Madrid, 1: ICM
    combat_df.loc['batch_id'] = batch_id
    combat_df = combat_df.transpose()
    madrid_combat_df = combat_df[combat_df['batch_id'] == '0']
    madrid_combat_df = madrid_combat_df.drop(columns=['batch_id'])
    icm_combat_df = combat_df[combat_df['batch_id'] == '1']
    icm_combat_df = icm_combat_df.drop(columns=['batch_id'])
    
    icm_combat_df.to_csv("/home/tachennf/Documents/delta-rad/data/ICM_0.35T/combat_csv/"+table, index=True)
    madrid_combat_df.to_csv("/home/tachennf/Documents/delta-rad/data/Madrid/combat_csv/"+table, index=True)
    
```
