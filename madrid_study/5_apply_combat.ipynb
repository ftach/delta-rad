```python
# columns = all participants
# rows = all features
madrid_simu = pd.read_csv(madrid_path+'SIMU_GTV_extracted_features_filtered.csv', index_col=0)
madrid_simu = madrid_simu.transpose()
madrid_simu.head()
```

```python
icm_simu = pd.read_csv(icm_path+'simu_mridian_gtv.csv', index_col=0)
icm_simu = icm_simu.transpose()
icm_simu.head()
```

### Join data

```python
simu = pd.concat([madrid_simu, icm_simu], axis=1)
simu.head()
```

### Remove NaN before Combat
